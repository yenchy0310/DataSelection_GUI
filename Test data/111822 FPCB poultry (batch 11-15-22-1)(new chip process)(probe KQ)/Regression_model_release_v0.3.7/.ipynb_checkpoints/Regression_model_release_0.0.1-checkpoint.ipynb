## AMS compensation process (for 4R_7341 system)

```python
def compensation(data, std_data):
        
    data['410nm #1_comp'] = data['410nm #1']* (std_data['410nm #2'][0]/data['410nm #2'])
    data['440nm #1_comp'] = data['440nm #1']* (std_data['440nm #2'][0]/data['440nm #2'])
    data['470nm #1_comp'] = data['470nm #1']* (std_data['470nm #2'][0]/data['470nm #2'])
    data['510nm #1_comp'] = data['510nm #1']* (std_data['510nm #2'][0]/data['510nm #2'])    
    data['550nm #1_comp'] = data['550nm #1']* (std_data['550nm #2'][0]/data['550nm #2'])
    data['583nm #1_comp'] = data['583nm #1']* (std_data['583nm #2'][0]/data['583nm #2'])
    data['620nm #1_comp'] = data['620nm #1']* (std_data['620nm #2'][0]/data['620nm #2'])
    data['670nm #1_comp'] = data['670nm #1']* (std_data['670nm #2'][0]/data['670nm #2'])
    data['900nm #1_comp'] = data['900nm #1']* (std_data['900nm #2'][0]/data['900nm #2'])
    
    data['410nm #2_comp'] = data['410nm #2']* (std_data['410nm #2'][0]/data['410nm #2'])
    data['440nm #2_comp'] = data['440nm #2']* (std_data['440nm #2'][0]/data['440nm #2'])
    data['470nm #2_comp'] = data['470nm #2']* (std_data['470nm #2'][0]/data['470nm #2'])
    data['510nm #2_comp'] = data['510nm #2']* (std_data['510nm #2'][0]/data['510nm #2'])    
    data['550nm #2_comp'] = data['550nm #2']* (std_data['550nm #2'][0]/data['550nm #2'])
    data['583nm #2_comp'] = data['583nm #2']* (std_data['583nm #2'][0]/data['583nm #2'])
    data['620nm #2_comp'] = data['620nm #2']* (std_data['620nm #2'][0]/data['620nm #2'])
    data['670nm #2_comp'] = data['670nm #2']* (std_data['670nm #2'][0]/data['670nm #2'])
    data['900nm #2_comp'] = data['900nm #2']* (std_data['900nm #2'][0]/data['900nm #2'])
    
    data['410nm #1_mv_comp'] = data['410nm #1_mv']* (std_data['410nm #2_mv'][0]/data['410nm #2_mv'])
    data['440nm #1_mv_comp'] = data['440nm #1_mv']* (std_data['440nm #2_mv'][0]/data['440nm #2_mv'])
    data['470nm #1_mv_comp'] = data['470nm #1_mv']* (std_data['470nm #2_mv'][0]/data['470nm #2_mv'])
    data['510nm #1_mv_comp'] = data['510nm #1_mv']* (std_data['510nm #2_mv'][0]/data['510nm #2_mv'])    
    data['550nm #1_mv_comp'] = data['550nm #1_mv']* (std_data['550nm #2_mv'][0]/data['550nm #2_mv'])
    data['583nm #1_mv_comp'] = data['583nm #1_mv']* (std_data['583nm #2_mv'][0]/data['583nm #2_mv'])
    data['620nm #1_mv_comp'] = data['620nm #1_mv']* (std_data['620nm #2_mv'][0]/data['620nm #2_mv'])
    data['670nm #1_mv_comp'] = data['670nm #1_mv']* (std_data['670nm #2_mv'][0]/data['670nm #2_mv'])
    data['900nm #1_mv_comp'] = data['900nm #1_mv']* (std_data['900nm #2_mv'][0]/data['900nm #2_mv'])
    
    data['410nm #2_mv_comp'] = data['410nm #2_mv']* (std_data['410nm #2_mv'][0]/data['410nm #2_mv'])
    data['440nm #2_mv_comp'] = data['440nm #2_mv']* (std_data['440nm #2_mv'][0]/data['440nm #2_mv'])
    data['470nm #2_mv_comp'] = data['470nm #2_mv']* (std_data['470nm #2_mv'][0]/data['470nm #2_mv'])
    data['510nm #2_mv_comp'] = data['510nm #2_mv']* (std_data['510nm #2_mv'][0]/data['510nm #2_mv'])    
    data['550nm #2_mv_comp'] = data['550nm #2_mv']* (std_data['550nm #2_mv'][0]/data['550nm #2_mv'])
    data['583nm #2_mv_comp'] = data['583nm #2_mv']* (std_data['583nm #2_mv'][0]/data['583nm #2_mv'])
    data['620nm #2_mv_comp'] = data['620nm #2_mv']* (std_data['620nm #2_mv'][0]/data['620nm #2_mv'])
    data['670nm #2_mv_comp'] = data['670nm #2_mv']* (std_data['670nm #2_mv'][0]/data['670nm #2_mv'])
    data['900nm #2_mv_comp'] = data['900nm #2_mv']* (std_data['900nm #2_mv'][0]/data['900nm #2_mv'])
```

## Overview data

```python
'''Visualize 4D data, x=Humidity, y=Temperautre, z=channel, color=ppm'''
def plot_4D(calibration_data, blindtest_data, channels):
    data = []
    for channel in channels:
        for name, df_data in calibration_data.items():
            trace = go.Scatter3d(x = df_data['Humidity'],
                                 y = df_data['Temperature'],
                                 z = df_data[channel],
                                 mode='markers',
                                 name=name,
                                 marker=dict(size=2,
                                             symbol='circle',
                                             opacity=0.5,
                                             color=df_data['ppm'],
                                             colorscale=[[0, 'rgb(255,160,122)'], [1.0, 'rgb(128,0,0)']])
                                             )
            data.append(trace)

        for name, df_data in blindtest_data.items():
            trace_1 = go.Scatter3d(x = df_data['Humidity'],
                                   y = df_data['Temperature'],
                                   z = df_data[channel],
                                   mode='markers',
                                   name=name,
                                   marker=dict(size=2,
                                               symbol='circle',
                                               opacity=0.5,
                                               color=df_data['ppm'],
                                               colorscale=[[0, 'rgb(192,192,192)'], [1.0, 'rgb(0,0,0)']])
                                               )
            data.append(trace_1)
            
    # convert channels list to string        
    channels = ', '.join(channels)    
    layout = go.Layout(title = '4D '+ channels,
                       scene = dict(xaxis = dict(title='Humidity'),
                                    zaxis = dict(title=channel),
                                    yaxis = dict(title='Temperature')),
                                    margin=dict(l=50, r=50, b=50, t=50))
        
    fig = go.Figure(data=data, layout=layout)
    cf.iplot(fig)
```

## Modeling

```python
from sklearn.linear_model import Ridge
```

```python
# setting channel
degree = 3
humidity_step = ['RH30%', 'RH40%', 'RH50%', 'RH60%', 'RH70%', 'RH80%']
step = humidity_step

'''4R_7341 calibration channel'''
channel = ['410nm #1', '440nm #1', '470nm #1', 
           '510nm #1', '550nm #1', '583nm #1', 
           '620nm #1', '670nm #1', '900nm #1']

channel_comp = ['410nm #1_comp', '440nm #1_comp', '470nm #1_comp', 
                '510nm #1_comp', '550nm #1_comp', '583nm #1_comp', 
                '620nm #1_comp', '670nm #1_comp', '900nm #1_comp']

channel_mv = ['410nm #1_mv', '440nm #1_mv', '470nm #1_mv', 
              '510nm #1_mv', '550nm #1_mv', '583nm #1_mv', 
              '620nm #1_mv', '670nm #1_mv', '900nm #1_mv']

channel_mv_comp = ['410nm #1_mv_comp', '440nm #1_mv_comp', '470nm #1_mv_comp', 
                   '510nm #1_mv_comp', '550nm #1_mv_comp', '583nm #1_mv_comp', 
                   '620nm #1_mv_comp', '670nm #1_mv_comp', '900nm #1_mv_comp']

# channels = channel + channel_comp + channel_mv + channel_mv_comp 
channels = channel_mv_comp
```

### Ridge regression

```python
loss_list = []
loss_each_step_list = []
for i in channels:
    ridge = modeling(channel=i,
                      x_train=x_train, 
                      y_train=y_train, 
                      x_test=x_test, 
                      y_test=y_test, 
                      model_name=Ridge(alpha=10), 
                      degree=3,
                      step=step,
                      output_modify=1,
                      asImage=True)
    pred, loss = ridge.regression()
    loss_list.append(loss)
#     ridge.loss()
    loss_each_step = ridge.loss_each_step()
    loss_each_step_list.append(loss_each_step)
    ridge.loss_noRH30()
    ridge.plot()
#     ridge.coef()

# plot each channel RMSE in different RH 
df_loss_each_step = pd.DataFrame(index=channels, columns=humidity_step, data=loss_each_step_list)
df_loss_each_step.iplot(kind='bar', yTitle='RMSE', title='RMSE in different RH', barmode='stack')
df_loss_each_step.to_csv('RMSE_in_different_RH.csv')

# plot each channel average RMSE 
df_loss = pd.DataFrame(index=channels, data=loss_list, columns=['RMSE test'])
df_loss.iplot(kind='bar', yTitle='RMSE', title='RMSE test')
df_loss.to_csv('RMSE_test_average.csv')
```
