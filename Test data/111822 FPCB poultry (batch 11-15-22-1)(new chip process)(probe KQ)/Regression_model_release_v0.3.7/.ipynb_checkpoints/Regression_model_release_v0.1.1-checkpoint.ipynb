```python
def white_card_regression(temp, white_intercept, white_coef):
    white_card_std = white_intercept + white_coef*temp
    return white_card_std
```

```python
def createWhiteCardStdColumn(data):
    white_std_list = []
    for temp in data['Temperature']:        
        white_card_std = white_card_regression(temp, white_intercept, white_coef)
        white_std_list.append(float(white_card_std)) 
    data['white card std'] = white_std_list
```

```python
def calculateEachSurfaceSignal(coef, x0, x1, degree): 
    '''Caculate the signal of each data pair of temperature and humidity  '''
    
    if degree ==1 :
        term = [1, x0, x1]
    
    elif degree == 2:
        term = [1, x0, x1, x0**2, x0*x1, x1**2]
        
    elif degree == 3:
        term = [1, x0, x1, x0**2, x0*x1, x1**2, x0**3, x0**2*x1, x0*x1**2, x1**3]
        
    elif degree == 4:
         term = [1, x0, x1, x0**2, x0*x1, x1**2, x0**3, x0**2*x1, x0*x1**2, x1**3, x0**4, x0**3*x1, x0**2*x1**2, x0*x1**3, x1**4]
                                                                                                                            
    else:
        print('Check degree value') 
        
    signal = np.inner(coef, term)
#     print(signal)
    return signal
```

```python
def interpolation(signal, signal_0ppm, signal_10ppm, signal_60ppm):
    '''predict ppm by interpolation algorithm'''
    
    if (signal_0ppm < signal_60ppm): # positive response
    
        if (signal>signal_0ppm)&(signal<signal_10ppm):
            ppm = abs(10-0)/abs(signal_10ppm-signal_0ppm)*abs((signal)-signal_0ppm) #interpolation
            return ppm

        elif (signal>signal_10ppm)&(signal<signal_60ppm):
            ppm = abs(60-10)/abs(signal_60ppm-signal_10ppm)*abs((signal)-signal_10ppm)+10 #interpolation
            return ppm

        elif (signal > signal_60ppm):
            ppm = abs(60-10)/abs(signal_60ppm-signal_10ppm)*abs((signal)-signal_60ppm)+60
            return ppm

        else:
            ppm = (-10)/(signal_0ppm-signal_10ppm)*((signal)-signal_0ppm)
            return ppm
        
    else: # negative response
        
        if (signal<signal_0ppm)&(signal>signal_10ppm):
            ppm = abs(10-0)/abs(signal_10ppm-signal_0ppm)*abs((signal)-signal_0ppm) #interpolation
            return ppm

        elif (signal<signal_10ppm)&(signal>signal_60ppm):
            ppm = abs(60-10)/abs(signal_60ppm-signal_10ppm)*abs((signal)-signal_10ppm)+10 #interpolation
            return ppm

        elif (signal < signal_60ppm):
            ppm = abs(60-10)/abs(signal_60ppm-signal_10ppm)*abs((signal)-signal_60ppm)+60
            return ppm

        else:
            ppm = (-10)/(signal_0ppm-signal_10ppm)*((signal)-signal_0ppm)
            return ppm
    
```

```python
def FFT_transform(data):
    
    '''Fourier Transform'''
    
    FFT_channel = ['410nm #2_mv', '440nm #2_mv', '470nm #2_mv', '510nm #2_mv', '550nm #2_mv', '583nm #2_mv', '620nm #2_mv', '670nm #2_mv']
    columns_real = ['410nm #2_mv_FFT_real', '440nm #2_mv_FFT_real', '470nm #2_mv_FFT_real', '510nm #2_mv_FFT_real', '550nm #2_mv_FFT_real', '583nm #2_mv_FFT_real', '620nm #2_mv_FFT_real', '670nm #2_mv_FFT_real']
    columns_imag = ['410nm #2_mv_FFT_imag', '440nm #2_mv_FFT_imag', '470nm #2_mv_FFT_imag', '510nm #2_mv_FFT_imag', '550nm #2_mv_FFT_imag', '583nm #2_mv_FFT_imag', '620nm #2_mv_FFT_imag', '670nm #2_mv_FFT_imag']
    columns_amp = ['410nm #2_mv_FFT_amp', '440nm #2_mv_FFT_amp', '470nm #2_mv_FFT_amp', '510nm #2_mv_FFT_amp', '550nm #2_mv_FFT_amp', '583nm #2_mv_FFT_amp', '620nm #2_mv_FFT_amp', '670nm #2_mv_FFT_amp']
    FFT_real = np.fft.fft(data[FFT_channel]).real
    FFT_imag = np.fft.fft(data[FFT_channel]).imag
    FFT_amp = np.abs(np.fft.fft(data[FFT_channel]))
    df_FFT_real = pd.DataFrame(FFT_real, columns=columns_real)
    df_FFT_imag = pd.DataFrame(FFT_imag, columns=columns_imag)
    df_FFT_amp = pd.DataFrame(FFT_amp, columns=columns_amp)
    data = pd.concat([data, df_FFT_real, df_FFT_imag, df_FFT_amp], axis=1, copy=True)
    return data
```

```python
#setting humidity range
humidity_data_dict = {                                         
#         '30%~50%': (50, 200),        #30%, 30%, 40%, 50%
#         '40%~70%': (150, 300),     #40%, 50%, 60%, 70%
    
        '0%~70%': (50, 300),     
    
#           'humidity 4': (0, 1500),
#           'humidity 5': (1000, 2000),
#             'humidity 6': (1500, 2500),
#             'humidity 7': (2000, 3000),
#             'humidity 8': (2500, 3500),    
             
                  }

# implement humidity range loop
for C, D in humidity_data_dict.values():
    # load training data
    dk_n20C = df_n20C[C:D].copy().reset_index(drop=True)
    dk_n10C = df_n10C[C:D].copy().reset_index(drop=True)
    dk_0C = df_0C[C:D].copy().reset_index(drop=True)
    dk_10C = df_10C[C:D].copy().reset_index(drop=True)
#     dk_20C = df_20C[C:D].copy().reset_index(drop=True)
#     dk_30C = df_30C[C:D].copy().reset_index(drop=True)
#     dk_40C = df_40C[C:D].copy().reset_index(drop=True)
#     dk_50C = df_50C[C:D].copy().reset_index(drop=True)
 
    # Load testing data
    dk_test_0C = df_test_0C[C:D].copy().reset_index(drop=True)
#     dk_test_0C_2 = df_test_0C_2[C:D].copy().reset_index(drop=True)
    dk_test_n10C = df_test_n10C[C:D].copy().reset_index(drop=True)
#     dk_test_10C = df_test_10C[C:D].copy().reset_index(drop=True)
#     dk_test_20C = df_test_20C[C:D].copy().reset_index(drop=True)
#     dk_test_20C_2 = df_test_20C_2[C:D].copy().reset_index(drop=True)
#     dk_test_20C_3 = df_test_20C_3[C:D].copy().reset_index(drop=True)
#     dk_test_40C = df_test_40C[C:D].copy().reset_index(drop=True)
    
#     dk_test_20C_4 = df_test_20C_4.copy().reset_index(drop=True)
#     dk_test_20C_5 = df_test_20C_5.copy().reset_index(drop=True)
#     dk_test_20C_6 = df_test_20C_6.copy().reset_index(drop=True)
#     dk_test_20C_7 = df_test_20C_7.copy().reset_index(drop=True)
    
#     dk_test_25C_1 = df_test_25C_1.copy().reset_index(drop=True)
#     dk_test_25C_2 = df_test_25C_2.copy().reset_index(drop=True)
    
#     dk_test_15C_1 = df_test_15C_1.copy().reset_index(drop=True)
#     dk_test_15C_2 = df_test_15C_2.copy().reset_index(drop=True)

    #setting temperature range
    '''dict = {temp. 1 : ([training data1, data2], [blind test data1, data2], white_card_std data)}'''
    
    temp_data_dict = {  
             # paper based setting
            'temp. 1': ([dk_n20C, dk_n10C, dk_0C, dk_10C], [dk_test_n10C, dk_test_0C], dk_n10C),
        
#             'temp. 1-2': ([dk_n10C, dk_0C, dk_10C], [dk_test_0C], dk_n10C),
#             'temp. 2': ([dk_30C, dk_40C, dk_50C],[dk_test_40C], dk_n10C),
#             'temp. 3': ([dk_30C, dk_40C, dk_50C], [dk_40C], dk_n10C),
        
#         'temp. 4': ([dk_0C, dk_10C, dk_20C, dk_30C], [dk_test_15C_1, dk_test_15C_2], dk_10C),
#         'temp. 5': ([dk_30C, dk_40C, dk_50C], [dk_test_30C], dk_10C),
        
        
#         'temp. 3': ([dk_0C, dk_10C, dk_20C], [dk_test_20C], dk_10C),
#         'temp. 4': ([dk_20C, dk_30C, dk_40C], [dk_test_20C], dk_10C),
#         'temp. 5': ([dk_30C, dk_40C, dk_50C], [dk_30C], dk_10C),
        
#         'temp. 5': ([dk_0C, dk_10C, dk_20C, dk_30C, dk_40C, dk_50C], [dk_test_20C], dk_10C),
        
                      } 
    
    # implement temp. range loop
    
    for temp_data, temp_blind_test, k_white_card_std in temp_data_dict.values():
        
#                    
            # plot each model 4D data
#             module_4D.plot_4D_multi_model(temp_data, temp_blind_test, channel_list, sensor_number)
    
            # concate training data        
            df_all = pd.concat(temp_data).reset_index(drop=True)
            df_test = pd.concat(temp_blind_test).reset_index(drop=True)
            
            
            # 取0ppm資料
            dk_all_0ppm = df_all[df_all['ppm']<1].copy()
            dk_test_0ppm = df_test[df_test['ppm']<1].copy()
            
            # 取10ppm資料
            dk_all_10ppm = df_all[(df_all['ppm']>9)&(df_all['ppm']<11)].copy()
            dk_test_10ppm = df_test[(df_all['ppm']>9)&(df_all['ppm']<11)].copy()
            
            # 取60ppm資料
            dk_all_60ppm = df_all[(df_all['ppm']>59)].copy()
            dk_test_60ppm = df_test[(df_all['ppm']>59)].copy()
            
            dk_data_dict = {'0ppm':(dk_all_0ppm, dk_test_0ppm),
                            '10ppm':(dk_all_10ppm, dk_test_10ppm),
                            '60ppm':(dk_all_60ppm, dk_test_60ppm)}

            #training model for 0ppm, 10ppm, 60ppm surface
            
#             temp_range_surf_coef_list
            surf_coef_list = [] # store ppm surface coefficient
            for ppm, ppm_data in dk_data_dict.items():
                
                dk_data, dk_test = ppm_data
            
                y_train = dk_data[train_channel] + shift 
                x_train = dk_data[['Temperature', 'Humidity']]  

                y_test = dk_test[test_channel] 
                x_test = dk_test[['Temperature', 'Humidity']]  

#                 y_train = y_train + shift
                channels = [train_channel]

                ### Ridge regression
    #             step = humidity_step
                os.chdir(path)                
                loss_list = []
                loss_each_step_list = []
                for i in channels:
                    ridge = modeling(name=name,
                                     sensor_number=sensor_number,
                                     channel=i,
                                      x_train=x_train, 
                                      y_train=y_train, 
                                      x_test=x_test, 
                                      y_test=y_test, 
                                      model_name=model_name, 
                                      degree=degree,
                                      step=humidity_step,
                                      output_modify=output_modify,
                                      white_card_std=k_white_card_std,
                                      white_card_side=white_card_side,
                                      white_card_std_multiple=white_card_std_multiple,
                                      shift=shift,            
                                      multiple=multiple,
                                      ppm=ppm)
                    intercept, coef, TL, TU, HL, HU = ridge.regression()
#                     loss_list.append(loss)
#                     ridge.loss()
                #     loss_each_step = ridge.loss_each_step()    
#                     ridge.y_pred_scale_buffer(buffer=2)
#                     ridge.plot()
                    ridge.coef()
                    coefficient = np.insert(coef, 0, intercept)
                    surf_coef_list.append(coefficient)
#                   

            test_data = df_test[[test_channel, 'Temperature', 'Humidity', 'ppm']]
            signal_0ppm_list = []
            signal_10ppm_list = []
            signal_60ppm_list = []
            ppm_list = []
            for index, row in test_data.iterrows():
                signal, temperature, humidity, _ = row

                signal_0ppm = calculateEachSurfaceSignal(surf_coef_list[0], x0=temperature, x1=humidity, degree=degree)
                signal_0ppm_list.append(signal_0ppm)

                signal_10ppm = calculateEachSurfaceSignal(surf_coef_list[1], x0=temperature, x1=humidity, degree=degree)
                signal_10ppm_list.append(signal_10ppm)

                signal_60ppm = calculateEachSurfaceSignal(surf_coef_list[2], x0=temperature, x1=humidity, degree=degree)
                signal_60ppm_list.append(signal_60ppm)

            #     print(signal_0ppm, signal_10ppm, signal_60ppm)

                ppm = interpolation(signal, signal_0ppm, signal_10ppm, signal_60ppm) # for positive response
#                 ppm = interpolation_nagative_response(signal, signal_0ppm, signal_10ppm, signal_60ppm) # for negative response
                ppm_list.append(float(ppm))
                
            # RMSE
            rmse_test = np.sqrt(mean_squared_error(ppm_list, test_data['ppm'])).round(2)

            # Interpolation prediction result    
            predict_test_data = test_data.copy()
            predict_test_data['signal_0ppm'] = signal_0ppm_list
            predict_test_data['signal_10ppm'] = signal_10ppm_list
            predict_test_data['signal_60ppm'] = signal_60ppm_list
            predict_test_data['pred_ppm'] = ppm_list
            temp_min = predict_test_data['Temperature'].min().round(2)
            temp_max = predict_test_data['Temperature'].max().round(2)
            hum_min = predict_test_data['Humidity'].min().round(2)
            hum_max = predict_test_data['Humidity'].max().round(2)
            title = 'model:({}C~{}C) ({}%~{}%) ({}) (RMSE={})'.format(TL, TU, HL, HU, test_channel, rmse_test)
            predict_test_data.iplot(kind='scatter', y=['ppm', 'pred_ppm', 'Humidity', 'Temperature'], title=title)
            print(surf_coef_list[0])
            print(surf_coef_list[1])
            print(surf_coef_list[2])
            predict_test_data
            
            # save prediction result
            folderName = '{}_degree={}'.format(name, degree)
            savePath = os.path.join(os.getcwd(), folderName)
            filename = '({}C~{}C) ({}%~{}%) ({}) (RMSE={})'.format(TL, TU, HL, HU, train_channel, rmse_test)
            plt.plot(predict_test_data['Temperature'], 'r')
            plt.plot(predict_test_data['Humidity'], 'b')
            plt.plot(predict_test_data['ppm'], 'g')
            plt.plot(predict_test_data['pred_ppm'], 'orange')
            plt.ylabel('Temperature/ Humidity/ ppm')
            plt.grid(alpha=0.3)
            plt.legend(bbox_to_anchor=(0., 1.02, 1., .102),  loc=3, ncol=4, mode="expand")
            plt.title(filename, y=1.15, fontsize=10)
            plt.savefig(savePath + '/' + filename + '.png', dpi=200, bbox_inches='tight')
            plt.close() 
            plt.show()
```
